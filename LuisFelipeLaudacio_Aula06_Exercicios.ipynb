```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
dados = np.array(np.random.randint(1,20, size=10))
condicao = dados % 2 == 0

resultado = np.extract(condicao, dados)

print(f'Array completo: {dados}')
print(f'Array filtrado: {resultado}')
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
lista1 = np.array(np.random.randint(1,50, size=10))
lista2 = np.array(np.random.randint(1,50, size=10))

print(f'Lista 1: {lista1}')
print(f'Lista 2: {lista2}\n')

resultadoSoma = lista1 + lista2
resultadoSubtracao = lista1 - lista2
resultadoMultiplicacao = lista1 * lista2
resultadoDivisao = lista1 / lista2

print(f'Resultado soma: {resultadoSoma}')
print(f'Resultado subtração: {resultadoSubtracao}')
print(f'Resultado multiplicação: {resultadoMultiplicacao}')
print(f'Resultado divisão: {resultadoDivisao}')
```

# 03 - Indexação e Slicing

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
produto = pd.Series(["Sabonete"] * 12)
meses = pd.Series([f'Mês {i+1}' for i in range(12)])
preco = pd.Series(np.random.randint(10,100, size=12))

df_preco = pd.DataFrame({"Produto": produto, "Meses": meses, "Preço": preco})

df_preco

# Create a Figure
fig = plt.figure(figsize=(8,5))

x = df_preco['Meses']
y = df_preco['Preço']

# Criando o gráfico de linha
plt.plot(x, y, label='Variação', color='blue')

# Adicionando título e rótulos
plt.title('Variações de Preços')
plt.xlabel('Meses')
plt.ylabel('Preço')
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='y', alpha=0.7)

# Adicionando a legenda
plt.legend()

# Exibindo o gráfico
plt.show()
```

# 13 - Histograma

Crie um histograma para mostrar a distribuição de idades em um conjunto de dados fictício de 100 pessoas.

```python
nome = pd.Series(["Nome"] * 100)
idade = pd.Series(np.random.randint(5,80, size=100))

df_idade = pd.DataFrame({"Nome": nome, "Idade": idade})

# Get the variable to examine
dados = df_idade["Idade"]

# Create a Figure
fig = plt.figure(figsize=(10,4))

# Plot a histogram
plt.hist(dados, color='skyblue', edgecolor='black')

# Add titles and labels
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Qtd. Pessoas')
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='y', alpha=0.7)

# Show the figure
fig.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
# Dados fictícios
altura = pd.Series(np.linspace(1, 2, num=10))
peso = pd.Series(np.random.randint(5,80, size=10))

df_altura_peso = pd.DataFrame({"Peso": peso, "Altura": altura})

# Criando o gráfico de dispersão
sns.lmplot(data=df_altura_peso, x="Peso", y="Altura")

# Adicionando título e rótulos
plt.title('Gráfico de Dispersão com Linha de Regressão')
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='y', alpha=0.7)

# Exibindo o gráfico
plt.show()
```

# 18 - Boxplot

Crie um boxplot para visualizar a distribuição das idades em um conjunto de dados fictício de 50 pessoas.

```python
# Dados fictícios
nome = pd.Series(["Nome"] * 50)
idade = pd.Series(np.random.randint(1,80, size=50))

df_idade = pd.DataFrame({"Nome": nome, "Idade": idade})

# Criando o gráfico boxplot
sns.boxplot(data=df_idade, x="Nome", y="Idade", gap=.1)

# Adicionando título e rótulos
plt.title('Gráfico Boxplot')
plt.ylabel('Idade')
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='y', alpha=0.7)

# Exibindo o gráfico
plt.show()


```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
# Dados fictícios
idade = pd.Series(np.random.randint(1,80, size=15))
altura = pd.Series(np.linspace(1, 2, num=15))
peso = pd.Series(np.random.randint(5,80, size=15))

df_idade = pd.DataFrame({"Idade": idade, "Altura": altura, "Peso": peso})

# Criando o gráfico heatmap
sns.heatmap(df_idade, annot=True)

# Adicionando título
plt.title('Gráfico Heatmap')

# Exibindo o gráfico
plt.show()

```
